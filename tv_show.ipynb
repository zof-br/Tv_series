## **The Crown**


Photo: Cast in season 1

**The Crown** is a historical drama television series about the reign of Queen Elizabeth II, created and principally written by Peter Morgan and produced by Left Bank Pictures and Sony Pictures Television for Netflix. The first season was initially realeased in 2016, while the last one in 2023.

The series consists of six seasons, spanning almost six decades, beginning shortly before the wedding of Princess Elizabeth and Philip Mountbatten in 1947 and ending with the 2005 wedding of Prince Charles and Camilla Parker Bowles. 

**Seasons**

- The first season covers 1947 to 1955, including Winston Churchill resigning as prime minister and Elisabeth coronation.
- The second season covers 1956 to 1964, including the Suez Crisis, the resignations of Anthony Eden and Harold Macmillan as prime minister, the Profumo affair, and the births of Prince Andrew and Prince Edward.
- The third season covers 1964 to 1977, including Harold Wilson and Edward Heath's terms as prime minister, the introduction of Camilla Shand, and the Queen's Silver Jubilee.
- The fourth season covers 1979 to 1990, including Margaret Thatcher's term as prime minister and the marriage of Lady Diana Spencer and Charles, Prince of Wales.
- The fifth season covers 1991 to 1997 and focuses on the "War of the Waleses" and the resulting divorce of Charles and Diana, as well as the rise of the Al-Fayed family and the Queen's "annus horribilis" of 1992.
- The sixth season covers 1997 to 2005, during the premiership of Tony Blair. The first four episodes cover the death of Diana, Princess of Wales, and the final six the Golden Jubilee of Elizabeth II, the deaths of Princess Margaret and The Queen Mother, the early relationship of Prince William and Kate Middleton, and the wedding of Prince Charles and Camilla Parker Bowles.

### Prizes

The series won twenty-one Primetime Emmy Awards, including Outstanding Drama Series for its fourth season and seven awards for the cast.It also won the Golden Globe Award for Best Television Series - Drama twice, at the 74th and 78th ceremonies, with additional acting wins for Foy, Colman, Corrin, O'Connor, Anderson, and Debicki.

### Rotten Tomatoes reviews

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4, 5, 6])
reviews_rt = np.array([0.88, 0.89, 0.90, 0.95, 0.71, 0.56])
avg_rt = 0.81
plt.plot(x, reviews_rt)
plt.xlabel("Season")
plt.ylabel("Rating (0-1)")
plt.axhline(y=avg_rt, color="red", linestyle = "dashed", label="average")
plt.legend()
plt.title("Rotten Tomatoes reviews - The Crown")
plt.show()
```

### Metacritic reviews

```python
reviews_m = np.array([81, 87, 84, 86, 65, 61])
avg_m = 78
plt.plot(x, reviews_m)
plt.xlabel("Season")
plt.ylabel("Rating (0-100)")
plt.axhline(y=avg_m, color="red", linestyle = "dashed", label="average")
plt.legend()
plt.title("Metacritic reviews - The Crown")
plt.show()
```

```python
mean1 = np.mean(reviews_rt[:4])
mean2= np.mean(reviews_rt[4:])
mean3= np.mean(reviews_m[0:4])
mean4 = np.mean(reviews_m[4:])
change = 100*mean2/mean1
change2= 100*mean4/mean3
print("The drop in ratings for Rotten Tomatoes was {}% between seasons 2-4 and 5-6.".format(round(change,2)))
print("Meanwhile, for Metacritic tha change was {}%.".format(round(change2, 2)))
```
